```python
def giveLastNgamesPlayer(player_id, date, n, bat_or_bowl):
    '''
    Function to get last n games stats of a player before an input date.
    
    Input-
    1. player_id: id of the player to get historical data.
    2. date: date to look-back and get n games. Stats returned are before this input date.
    3. n: Number of historical games stats to return.
    4. bat_or_bowl: Kind of stats to return. {'bat': batting stats to return, 'bowl': bowling stats to return}
    
    Output-None
    
    Returns- dataframe having bowling/batting stats from last n games of a player before an input date. 
    The results are sorted by date.
    '''
    if bat_or_bowl == 'bat':
        df_topick = batsman_data
        id_col = 'batsman_id'
    else:
        df_topick = bowler_data
        id_col = 'bowler_id'
        
    return df_topick[(df_topick['match_dt']<date)&(df_topick[id_col]==float(player_id))]\
                .sort_values(by='match_dt', ascending=False).head(n)
```

```python
def fetch_bat_perf(player_ids , date , n):
    '''
    This function is intended to calculate the overall batting performance of the team

    Input-
    1. player_ids = a list of player ids in int/float format
    2. date = date upto which data is required in str format
    3. n = int value specifying how much of the last games
    '''

    team_contri = [] #for storing the contribution to the average total of the team
    bat_avg = [] #for storing average runs of each of the players
    strike_rt = [] #for storing average strike rate of each of the players

    #filling the lists with corresponding features
    for i in range(len(player_ids)):
        bat_avg.append(giveLastNgamesPlayer(player_ids[i], date, n, "bat")["runs"].sum()/n)
        strike_rt.append(giveLastNgamesPlayer(player_ids[i], date, n, "bat")["strike_rate"].sum()/n)

    #filling up the team contribution of each of the player
    total_runs = sum(bat_avg) #summing up the average scores of each of the players
    team_contri = [syna/total_runs for syna in bat_avg]

    min_contri , max_contri = (min(team_contri) , max(team_contri)) #max and min of team cintributions
    min_runs , max_runs = (min(bat_avg) , max(bat_avg)) #max and min of avg player runs
    min_sr , max_sr = (min(strike_rt) , max(strike_rt)) #max and min of strike rates

    # print(team_contri)
    # print(bat_avg)
    # print(strike_rt)

    #normalising each of the parameters
    # team_contri = [((syna - min_contri)/(max_contri-min_contri)) for syna in team_contri]
    bat_avg = [((syna - min_runs)/(max_runs - min_runs)) for syna in bat_avg]
    strike_rt = [((syna - min_sr)/(max_sr - min_sr)) for syna in strike_rt]

    # print(team_contri)
    # print(bat_avg)
    # print(strike_rt)

    #calculating the variance of each of the parameters
    var_contri = np.var(team_contri) #variance of team contributions
    var_runs = np.var(bat_avg) #variance of average runs of each of the players
    var_sr = np.var(strike_rt) #variance of strike rate

    c = 1/((1/((var_contri)**0.5)) + (1/((var_runs)**0.5)) + (1/((var_sr)**0.5))) #this is the normalising factor that will be used to calculate weights later

    #calculation of weights
    w_contri = c/((var_contri)**0.5) #weight for team contribution
    w_runs = c/((var_runs)**0.5) #weight for average runs of each of the players
    w_sr = c/((var_sr)**0.5) #weight for strike rate of players

    # print(w_contri)
    # print(w_runs)
    # print(w_sr)

    bat_perf = [] #for storing overall batting performance of each of the players

    #calculating the overall batting performance of each of the players by multiplying weights to each of the parameters
    for i in range(len(player_ids)):
        bat_perf.append((w_contri*team_contri[i]) + (w_runs*bat_avg[i]) + (w_sr*strike_rt[i]))

    # print(bat_perf)
    # print(sum(bat_perf)/len(bat_perf))
    
    return sum(bat_perf)/len(bat_perf)
```

```python
def fetch_bowl_perf(player_ids , date , n):
    '''
    This function is intended to calculate the overall bowling performance of the team

    Input-
    1. player_ids = a list of player ids in int/float format
    2. date = date upto which data is required in str format
    3. n = int value specifying how much of the last games
    '''

    eco_rt = [] #for storing the average economy of each of the players
    bowl_avg = [] #for storing average bowling average of each of the players
    strike_rt = [] #for storing average strike rate of each of the players

    #filling the lists with corresponding features
    for i in range(len(player_ids)):
        wkts = giveLastNgamesPlayer(player_ids[i], date, n, "bowl")["wicket_count"].sum()/n
        runs = giveLastNgamesPlayer(player_ids[i], date, n, "bowl")["runs"].sum()/n
        economy = giveLastNgamesPlayer(player_ids[i], date, n, "bowl")["economy"].sum()/n
        balls = giveLastNgamesPlayer(player_ids[i], date, n, "bowl")["balls_bowled"].sum()/n

        eco_rt.append(economy)

        if wkts == 0:
            bowl_avg.append(0)
            strike_rt.append(0)
        else:
            bowl_avg.append(runs/wkts)
            strike_rt.append(balls/wkts)

    
    min_eco , max_eco = (min(eco_rt) , max(eco_rt)) #max and min of economy rates
    min_avg , max_avg = (min(bowl_avg) , max(bowl_avg)) #max and min of player bowling averages
    min_sr , max_sr = (min(strike_rt) , max(strike_rt)) #max and min of strike rates

    # print(eco_rt)
    # print(bowl_avg)
    # print(strike_rt)

    #normalising each of the parameters
    for i in range(len(player_ids)):
        if eco_rt[i] == 0:
            continue
        else:
            eco_rt[i] = ((max_eco - eco_rt[i])/(max_eco-min_eco))
    
    for i in range(len(player_ids)):
        if bowl_avg[i] == 0:
            if strike_rt[i] == 0:
                continue
            else:
                strike_rt[i] = (max_sr - strike_rt[i])/(max_sr - min_sr)
        else:
            if strike_rt[i] == 0:
                bowl_avg[i] = (max_avg - bowl_avg[i])/(max_avg - min_avg)
            else:
                bowl_avg[i] = (max_avg - bowl_avg[i])/(max_avg - min_avg)
                strike_rt[i] = (max_sr - strike_rt[i])/(max_sr - min_sr)
                    
    # bowl_avg = [((max_avg - syna)/(max_avg - min_avg)) if syna != np.nan else np.nan for syna in bowl_avg]
    # strike_rt = [((max_sr - syna)/(max_sr - min_sr)) if syna != np.nan else np.nan for syna in strike_rt]

    # print(eco_rt)
    # print(bowl_avg)
    # print(strike_rt)

    #calculating the variance of each of the parameters
    var_eco = np.nanvar(eco_rt) #variance of economy rates
    var_avg = np.nanvar(bowl_avg) #variance of bowling average of each of the players
    var_sr = np.nanvar(strike_rt) #variance of strike rate

    # print(var_eco , " " , var_avg , " " , var_sr)

    c = 1/((1/((var_eco)**0.5)) + (1/((var_avg)**0.5)) + (1/((var_sr)**0.5))) #this is the normalising factor that will be used to calculate weights later

    #calculation of weights
    w_eco = c/((var_eco)**0.5) #weight for economy rates
    w_avg = c/((var_avg)**0.5) #weight for bowling average of each of the players
    w_sr = c/((var_sr)**0.5) #weight for strike rate of players

    # print(w_eco)
    # print(w_avg)
    # print(w_sr)

    bowl_perf = [] #for storing the overall bowling performance of each of the players

    #caluculating the overall bowling performance of each of the players by multiplying weights to each of the parameters
    for i in range(len(player_ids)):
        bowl_perf.append((w_eco*eco_rt[i]) + (w_avg*bowl_avg[i]) + (w_sr*strike_rt[i]))

    # print(bowl_perf)
    # print(sum(bowl_perf)/len(bowl_perf))
    
    return sum(bowl_perf)/len(bowl_perf)
```

```python
def fetch_team_eff(player_ids , date , n):
    '''
    This function is intended to calculate the effective size of the team

    Input-
    1. player_ids = a list of player ids in int/float format
    2. date = date upto which data is required in str format
    3. n = int value specifying how much of the last games
    '''
    
    runs = [] #storing runs of each player
    wickets = [] #storing wkts of each player

    #filling the data into the parameters
    for i in range(len(player_ids)):
        runs.append(giveLastNgamesPlayer(player_ids[i], row["match_dt"], n, "bat")["runs"].sum()/n)
        wickets.append(giveLastNgamesPlayer(player_ids[i], row["match_dt"], n, "bowl")["wicket_count"].sum()/n)
    
    # print(runs)
    # print(wickets)

    #total runs and wickets of the team
    runs_total = sum(runs)
    wkt_total = sum(wickets)

    # print(runs_total , '\n' , wkt_total)

    #team relative bat and ball performance for each player
    runs = [syna/runs_total for syna in runs]
    wickets = [syna/wkt_total for syna in wickets]

    # print(runs)
    # print(wickets)

    #calculating performance of each of the players
    team_perf = [sum(i)*0.5 for i in zip(runs , wickets)]

    # print(team_perf)

    H_scores = [syna*(np.log2(syna)) for syna in team_perf] #H score for each of the players

    # print(H_scores)

    H_score = -1 * np.nansum(H_scores) #overall H score of the team

    # print(H_score)

    ply_eff = 2**H_score #the effective no. of players in the team

    # print(ply_eff)
    return runs_total , wkt_total , ply_eff
```

```python
def fetch_stats(row , n):
    '''
    This function would return all the values that have been calculated so far

    Inputs-
    1. row- the row of the train dataset for which we wish to find the stats, type df
    2. n = how many of the previous games of type int
    3. team = team no. 1 or 2, type int
    '''
    #preaparing a list of team ids
    team1 = row["team1_roster_ids"].split(":")
    team2 = row["team2_roster_ids"].split(":")

    team1 = list(map(float , team1))
    team2 = list(map(float , team2))

    # print(team1)
    # print(team2)

    t1_bat_perf = fetch_bat_perf(team1 , row["match_dt"] , n) #batting performance of team 1
    t2_bat_perf = fetch_bat_perf(team2 , row["match_dt"] , n) #batting performance of team 2

    t1_bowl_perf = fetch_bowl_perf(team1 , row["match_dt"] , n) #bowling performnace of team 1
    t2_bowl_perf = fetch_bowl_perf(team2 , row["match_dt"] , n) #bowling performance of team 2

    # print()
    # print(t1_bat_perf , " " , t1_bowl_perf)
    # print(t2_bat_perf , " " , t2_bowl_perf)
    # print()

    #normalising the performance scores
    max_t1 = max(t1_bat_perf , t1_bowl_perf) #maximum of the 2 scores of team 1
    max_t2 = max(t2_bat_perf , t2_bowl_perf) #maximum of the 2 scores of team 2

    # print(max_t1 , " " , max_t2)
    # print()

    # if max_t1 != 0:
    #     t1_bat_perf , t1_bowl_perf = (t1_bat_perf/max_t1 , t1_bowl_perf/max_t1)
    # else:
    #     t1_bat_perf , t1_bowl_perf = (0 , 0)

    # if max_t2 != 0:
    #     t2_bat_perf , t2_bowl_perf = (t2_bat_perf/max_t2 , t2_bowl_perf/max_t2)
    # else:
    #     t2_bat_perf , t2_bowl_perf = (0 , 0)

    t1_bat_perf , t1_bowl_perf = (t1_bat_perf/max_t1 , t1_bowl_perf/max_t1)
    t2_bat_perf , t2_bowl_perf = (t2_bat_perf/max_t2 , t2_bowl_perf/max_t2)

    # print(t1_bat_perf , " " , t1_bowl_perf)
    # print(t2_bat_perf , " " , t2_bowl_perf)
    # print()
    
    #calculating the overall performance of the team over batting and bowling
    k_t1 = 0.5*(t1_bat_perf + t1_bowl_perf)
    k_t2 = 0.5*(t2_bat_perf + t2_bowl_perf)

    # print(k_t1 , " " , k_t2)
    # print()

    #we will now convert the overall performances to form factors using the following transformations
    k_t1 = 1 - k_t1
    k_t1 = 1 - k_t2

    # print(k_t1 , " " , k_t2)
    # print()

    wins_t1 = giveLastNgamesTeams(row["team1"] , row["match_dt"] , n , 1)
    wins_t2 = giveLastNgamesTeams(row["team2"] , row["match_dt"] , n , 2)

    # print(wins_t1)
    # print(wins_t2)
    # print()

    #for the next set of calculations to work, we need win = 1 for cases where a team won. Hence we need to flip the encoding for team 1
    for i in range(len(wins_t1)):
        if wins_t1[i] == 0:
            wins_t1[i] = 1
        else:
            wins_t1[i] = 0

    # print(wins_t1)
    # print(wins_t2)
    # print()

    #if the number of matches returned is less than 'n' , then we need to pad wins with 0 at the end
    wins_t1 = np.pad(wins_t1 , (0 , (n-len(wins_t1))) , 'constant' , constant_values = 0)
    wins_t2 = np.pad(wins_t2 , (0 , (n-len(wins_t2))) , 'constant' , constant_values = 0)

    # print(wins_t1)
    # print(wins_t2)
    # print()

    #from the performances, we will now calculate the form of the teams for each of the matches so far
    wins_t1 = [k_t1*syna for syna in wins_t1]
    wins_t2 = [k_t2*syna for syna in wins_t2]
    # for i in range(n):
    #     t1_form = ((k_t1**i)*wins_t1[i])/(k_t1**i)
    #     t2_form = ((k_t2**i)*wins_t2[i])/(k_t1**i)

    # print(wins_t1 , " " , wins_t2)
    # print()

    #caluculating total runs & wickets and the effective team size of both teams
    t1_total_runs , t1_total_wkts , t1_eff_ply = fetch_team_eff(team1 , row["match_dt"] , n)
    t2_total_runs , t2_total_wkts , t2_eff_ply = fetch_team_eff(team2 , row["match_dt"] , n)

    # print(t1_total_runs , " " , t1_total_wkts , " " , t1_eff_ply)
    # print(t2_total_runs , " " , t2_total_wkts , " " , t2_eff_ply)
    # print()

    return t1_total_runs , t1_total_wkts , t1_bat_perf , t1_bowl_perf , t1_eff_ply , np.sum(wins_t1) ,\
    t2_total_runs , t2_total_wkts , t2_bat_perf , t2_bowl_perf , t2_eff_ply , np.sum(wins_t2)
```

```python
def winpCrossLastn(team1_id, team2_id, date, n):
    '''
    Function to compute team1's win% against team2 from the current game in their past n encounters.
    
    Input-
    1. team1_id: ID of team1 to calculate win% of.
    2. team2_id: ID of team2 to calculate win% against.
    3: date: match date of the current game for which the feature is to be calculated.
    4. n: look-back window of games for both these teams.
    
    Output- None
    
    Returns- Float value denoting team1's win% against team2 in their past n games against each other.
    '''
    # filter out games where either team1_id is input team1 and team2_id is input team2, or where team2_id is input team1 and team1_id is input team2.
    # Also, match date is less than current games's input date, sort desc by date and get top n rows (games)
    df_rel = match_data[(match_data['match_dt']<date)&\
                      (((match_data['team1_id']==team1_id)&(match_data['team2_id']==team2_id))|((match_data['team1_id']==team2_id)&(match_data['team2_id']==team1_id)))]\
                        .sort_values(by='match_dt', ascending=False).head(n)
    win_count = df_rel[df_rel['winner_id']==team1_id].shape[0] # Counting number of rows (games) where winner is input team1.
    if win_count == 0:
        return 0
    return round(win_count*100/df_rel.shape[0], 2) # return Float denoting team1's win% against team2 in past n games rounded to 2 decimal places.
```

```python
def ground_advantage(row , n):
    '''
    This function calculates the ralative no. of matches won by team 1 against team 2 on a ground before a particular date

    Input-
    1. row: the row of the dataframe to which it is going to be applied to
    2. n = how many matches out of them to be picked, type int
    '''

    t1_df = match_data[((match_data["winner"] == row["team1"]) & (match_data["ground_id"] == row["ground_id"]) \
                        &(match_data["match_dt"] < row["match_dt"]))].sort_values(by='match_dt', ascending=False).head(n)
    t2_df = match_data[((match_data["winner"] == row["team2"]) & (match_data["ground_id"] == row["ground_id"]) \
                        &(match_data["match_dt"] < row["match_dt"]))].sort_values(by='match_dt', ascending=False).head(n)
    count_t1 = t1_df.shape[0]
    # print(count_t1)
    count_t2 = t2_df.shape[0]
    # print(count_t2)

    if count_t2 == 0:
        return count_t1
    else:
        return count_t1/count_t2
```
